```python
df_doctor['Qual_1'] = df_doctor['Qualification'].str.split(',').str[0]
df_doctor['Qual_2'] = df_doctor['Qualification'].str.split(',').str[1]
df_doctor['Qual_3'] = df_doctor['Qualification'].str.split(',').str[2]
df_doctor['Qual_1'].fillna("not available",inplace = True)
df_doctor['Qual_2'].fillna("not available",inplace = True)
df_doctor['Qual_3'].fillna("not available",inplace = True)
#splitting the qualification column as the doctors have multiple qualification
#filling the empty column if no data is present
df_doctor['Rating'].fillna('0%',inplace = True)
#filling the Rating Column with 0%
df_doctor['City'] = df_doctor['Place'].str.split(',').str[1]
df_doctor['Locality'] = df_doctor['Place'].str.split(',').str[0]
df_doctor['City'].fillna("not available",inplace = True)
df_doctor['Locality'].fillna("not available",inplace = True)
#splitting Place into City and Locality and filling the nan values
```

```python
# Extract years of experience
df_doctor["Experience"] = df_doctor["Experience"].str.split()
df_doctor["Experience"] = df_doctor["Experience"].str[0].astype("int")
df_doctor["Rating"] = df_doctor["Rating"].str[:-1].astype("int")
#spliiting Experience and Rating column and extracting the no. of years as int and Rating as int
```

```python
#creating a function to find the random_state coresponding to the maximum r2score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn import linear_model
def maxr2_score(regr,df_x,y):
    maxr_score=0
    for r_state in range(42,101):
        x_train,x_test,y_train,y_test=train_test_split(df_x,y,random_state=r_state,test_size=0.20)
        regr=linear_model.LinearRegression()
        regr.fit(x_train,y_train)
        y_predict=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_predict)
        print("r2score corresponing to",r_state,"is:",r2_scr)
        if r2_scr>maxr_score:
            maxr_score=r2_scr
            final_r_state=r_state
    print()
    print()
    print("max r2_score corresponding to ",final_r_state,"is",maxr_score) 
    return final_r_state
```
