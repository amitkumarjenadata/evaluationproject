```python
#removing skewness from above columns
for col in insurance_data.columns:
    if insurance_data.skew().loc[col]>0.55:
        insurance_data[col]=np.log1p(insurance_data[col])
```

```python
insurance_data.skew()
```

```python
#checking for outliers and removing them
from scipy.stats import zscore
z_score=abs(zscore(insurance_data))
print(insurance_data.shape)
insurance_data_final=insurance_data.loc[(z_score<3).all(axis=1)]
print(insurance_data_final.shape)
#data reduced from 1338,8 to 1334,8
```

```python
x=insurance_data_final.iloc[:,:-1]
x.head()
#dividing the dataset into input
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=ss.fit_transform(x)
#scaling input data
```

```python
y=insurance_data_final.iloc[:,-1]
y.head()
#output data
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#splitting the data
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=9,test_size=0.3,stratify=y)
```

```python
#import the algorithm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#creating an object of the algorithm
knn=KNeighborsClassifier(n_neighbors=6)
sv=SVC()
dt=DecisionTreeClassifier(random_state=6)
gnb=GaussianNB()
lr=LogisticRegression()
```

```python
#using GridssearchCV to find the best parameter
from sklearn.model_selection import GridSearchCV
neighbors={"n_neighbors":range(1,30)}
gknn=GridSearchCV(knn,neighbors,cv=10)
gknn.fit(x,y)
gknn.best_params_
```

```python
#using GridsearchCV to find the best parameter
grid_param={"random_state":range(0,20),"criterion":['gini','entropy']}
Gdt=GridSearchCV(dt,param_grid=grid_param,scoring='accuracy',cv=10)
Gdt.fit(x,y)
Gdt.best_params_
```

```python
#using GridsearchCV to find the best parameter
kernel={"kernel":["linear","rbf","poly"]}
Gsv=GridSearchCV(sv,kernel,cv=10)
Gsv.fit(x,y)
Gsv.best_params_
```

```python
#the best parameters are provided for hyperparameter tuning
knn=KNeighborsClassifier(n_neighbors=3)
sv=SVC(kernel="rbf")
dt=DecisionTreeClassifier(random_state=1,criterion="entropy")
gnb=GaussianNB()
lr=LogisticRegression()
```

```python
models=[]
models.append(("KNeighborsClassifier",knn))
models.append(("SVC",sv))
models.append(("DecisionTreeClassifier",dt))
models.append(("GaussianNB",gnb))
models.append(("LogisticRegression",lr))
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
```

```python
Model=[]
score=[]
cvs=[]
roc_score=[]
for name,model in models:
    print('*********************',name,'***************************')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    AS=accuracy_score(y_test,pre)
    print("accuracy_score",AS)
    score.append(AS*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring="accuracy").mean()
    print("cross_val_score=",sc)
    cvs.append(sc*100)
    false_positive_rate,true_positive_rate,threshold=roc_curve(y_test,pre)
    roc_auc=auc(false_positive_rate,true_positive_rate)
    print("roc_auc_score=",roc_auc)
    roc_score.append(roc_auc*100)
    print('\n')
    print('classification_report\n',classification_report(y_test,pre))
    print("\n")
    cm=confusion_matrix(y_test,pre)
    print(cm)
    print('\n')
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(false_positive_rate,true_positive_rate,label='AUC=%0.2f'%roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('True positive rate')
    plt.xlabel('False positive rate')
    print('\n\n')
```

```python
result=pd.DataFrame({'Model':Model,'Accuracy_score':score,'Cross_Val_Score':cvs,'roc_auc_curve':roc_score})
result
```
