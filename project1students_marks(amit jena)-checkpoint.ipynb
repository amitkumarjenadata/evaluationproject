```python
from scipy.stats import zscore
z_score=abs(zscore(student_marks.iloc[:,3:11]))
print(student_marks.shape)
student_marks_final=student_marks.iloc[:,3:11].loc[(z_score<3).all(axis=1)]
print(student_marks_final.shape)
#outliers present the column has been removed the final shape has been reduced from (10,11) to (10,8)
```

```python
#dataset containing only the marks of the students
student_marks_final
```

```python
#none of the data is skewed
student_marks_final.skew()
```

```python
#removing skewness
import numpy as np
for col in student_marks_final.columns:
    if student_marks_final.skew().loc[col]>0.55:
        student_marks_final[col]=np.log1p(student_marks_final[col])
```
